```python
from google.colab import files
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

# Upload manual da minha imagem
uploaded = files.upload()
```

```python
# Função para converter RGB em Cinza
def rgb_to_gray(img_array):
    # Pega os canais R, G, B
    R = img_array[:, :, 0]
    G = img_array[:, :, 1]
    B = img_array[:, :, 2]
    # Aplica a formula de cinza
    gray = (0.299 * R + 0.587 * G + 0.114 * B).astype(np.uint8)
    return gray
```

```python
# Função para binarizar direto da colorida
def color_to_binary(img_array, threshold=128):
    # Média simples (somar e dividir) dos canais R, G, B
    intensity = (img_array[:, :, 0].astype(int) +
                 img_array[:, :, 1].astype(int) +
                 img_array[:, :, 2].astype(int)) // 3
    # Threshold: maior que 128 é branco, caso contrário é preto
    binary = np.where(intensity > threshold, 255, 0).astype(np.uint8)
    return binary
```

```python
# Função de processamento
def process_image(img_path, threshold=128):
    img = Image.open(img_path)
    img_array = np.array(img)

    # Escala de cinza
    gray = rgb_to_gray(img_array)

    # Binarização direto da colorida
    binary = color_to_binary(img_array, threshold)

    # Mostrar resultados
    plt.figure(figsize=(15,5))

    plt.subplot(1,3,1)
    plt.imshow(img_array)
    plt.title("Colorida")
    plt.axis("off")

    plt.subplot(1,3,2)
    plt.imshow(gray, cmap="gray")
    plt.title("Escala de Cinza (0–255)")
    plt.axis("off")

    plt.subplot(1,3,3)
    plt.imshow(binary, cmap="gray")
    plt.title("Binarizada (Preto/Branco)")
    plt.axis("off")

    plt.show()

# Executar o processo com a imagem enviada
process_image(list(uploaded.keys())[0])
```
